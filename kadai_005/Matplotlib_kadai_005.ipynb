```python
import numpy as np
import matplotlib.pyplot as plt

# 2×3のサブプロットを作成
fig, axes = plt.subplots(2, 3, tight_layout=True)

# 1つ目のグラフ（折れ線グラフ）
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)

# 2つ目のグラフ（sin関数）
x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)
axes[0, 1].plot(x1, y1)

# 3つ目のグラフ（ヒストグラム）
x2 = np.random.normal(15, 5, 2000)  # ここで定義
axes[0, 2].hist(x2)

# 4つ目のグラフ（散布図）
x3 = np.random.rand(100)
y3 = np.random.rand(100)
axes[1, 0].scatter(x3, y3)

# グラフを表示
plt.show()
```
